# Yearly data

On garde une première base de données qui ne contient que les pays avec aucune valeur manquante (un total de 67 pays et une longueur total de 30 ans), et une deuxième base de données qui contient les pays avec au plus 3 valeurs manquantes qui se trouve au début de la série et que l'on supprimer pour un total de 69 pays et une longueur total de 27 ans. 

```python
GDP_yearly = pd.read_csv('../Data/GDP_yearly.csv')
```

```python
GDP_yearly.rename(columns={GDP_yearly.columns[0] : 'Year'}, inplace=True)
```

```python
GDP_yearly = GDP_yearly.iloc[1:]
```

```python
GDP_yearly.head()
```

```python
missing_values = GDP_yearly.iloc[:, 1:].isnull().sum()
```

```python
missing_values
```

```python
# Colonnes avec entre 1 et 3 valeurs manquantes (inclus)
cols_with_some_missing = missing_values[(missing_values >= 1) & (missing_values <= 3)]
print("Colonnes avec entre 1 et 3 valeurs manquantes :")
print(cols_with_some_missing)

# Nombre de colonnes sans aucune valeur manquante
num_cols_no_missing = (missing_values == 0).sum()
print(f"\nNumber of countries without missing values : {num_cols_no_missing}")
```

```python
# Afficher les colonnes du DataFrame avec entre 1 et 3 valeurs manquantes
df_with_some_missing = GDP_yearly[cols_with_some_missing.index]
print("Colonnes avec entre 1 et 3 valeurs manquantes :")
df_with_some_missing
```

```python
# Sélectionner les colonnes avec 0 valeurs manquantes
cols_with_no_missing = missing_values[missing_values == 0]
```

```python
GDP_yearly_length_30 = GDP_yearly[cols_with_no_missing.index]
```

```python
# Sélectionner la première colonne (les dates)
dates_column = GDP_yearly.iloc[:, 0]

# Filtrer les colonnes avec 0 valeurs manquantes et ajouter la première colonne
GDP_yearly_length_30 = pd.concat([dates_column, GDP_yearly[cols_with_no_missing.index]], axis=1)
```

```python
# Mettre la première colonne (les dates) en indice
GDP_yearly_length_30.set_index(dates_column.name, inplace=True)
```

```python
# Colonnes avec entre 1 et 3 valeurs manquantes (inclus)
cols_with_some_missing = missing_values[missing_values <= 3]
print("Colonnes avec entre 1 et 3 valeurs manquantes :")
print(cols_with_some_missing)
```

```python
GDP_yearly_length_27 = pd.concat([dates_column, GDP_yearly[cols_with_some_missing.index]], axis=1).iloc[3:]
# Mettre la première colonne (les dates) en indice
GDP_yearly_length_27.set_index(dates_column.name, inplace=True)
```

```python
GDP_yearly_length_30.to_csv('GDP_yearly_length_30', index = True)
GDP_yearly_length_27.to_csv('GDP_yearly_length_27', index = True)
```

```python
GDP_yearly_length_30
```

## Data normalization

### Normalisation des données entre 0 et 1, avec calcul du min et du max pour l'ensemble des pays

```python
# Normalisation des données
GDP_yearly_27_normalized = (GDP_yearly_length_27 - GDP_yearly_length_27.min().min()) / (GDP_yearly_length_27.max().max() - GDP_yearly_length_27.min().min())
GDP_yearly_27_normalized.to_csv('GDP_yearly_27_global_normalized', index=False)

# Normalisation des données
GDP_yearly_30_normalized = (GDP_yearly_length_30 - GDP_yearly_length_30.min().min()) / (GDP_yearly_length_30.max().max() - GDP_yearly_length_30.min().min())
GDP_yearly_30_normalized.to_csv('GDP_yearly_30_global_normalized', index=False)
```

### Normalisation des données entre 0 et 1, avec calcul du min et du max pour chaque colonne 

```python
# Sélectionner les colonnes numériques (ignorer la colonne 'Année')
cols_to_normalize = GDP_yearly_length_30.columns
```

```python
# Normalisation min-max
GDP_yearly_length_30[cols_to_normalize] = GDP_yearly_length_30[cols_to_normalize].apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Afficher le DataFrame normalisé
print("DataFrame avec les données normalisées entre 0 et 1 :")
print(GDP_yearly_length_30)
```

```python
GDP_yearly_length_30.to_csv('GDP_yearly_length_30_normalized', index=True)
```

```python
# Sélectionner les colonnes numériques (ignorer la colonne 'Année')
cols_to_normalize = GDP_yearly_length_27.columns

# Normalisation min-max
GDP_yearly_length_27[cols_to_normalize] = GDP_yearly_length_27[cols_to_normalize].apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Afficher le DataFrame normalisé
print("DataFrame avec les données normalisées entre 0 et 1 :")
print(GDP_yearly_length_27)
```

```python
GDP_yearly_length_27.to_csv('GDP_yearly_length_27_normalized', index=True)
```
